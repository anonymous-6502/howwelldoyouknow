```python
import numpy as np
```

```python
def load_npy(path):
    with open(path, 'rb') as fp:
        data = np.load(fp)
        
    print(data.shape)
    return data
```

```python
pr_giga = load_npy('./gigaword_pearson.npy')
pr_cnndm = load_npy('./cnndm_pearson.npy')
pr_xsum = load_npy('./xsum_pearson.npy')
```

```python
labels_giga = ['R1', 'R2', 'RL', 'MS', 'BS']
labels_cnndm = labels_giga + ['HP']
labels_xsum = labels_giga
```

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
viridis = matplotlib.cm.get_cmap('viridis')

fig, ax = plt.subplots(1, 3, figsize=(10, 4))
cbar_ax = fig.add_axes([1, 0.135, .02, 0.75])

# Giga
b = sns.heatmap(pr_giga, xticklabels=labels_giga, yticklabels=labels_giga, linewidths=0.1, 
            annot=True, ax=ax[0], cmap=viridis, vmin=0, vmax=1, cbar=False);
ax[0].set_title('(a) Gigaword', fontsize=15)
b.set_xticklabels(labels_giga, size = 15)
b.set_yticklabels(labels_giga, size = 15)

# CNN/DM
b = sns.heatmap(pr_cnndm, xticklabels=labels_cnndm, yticklabels=labels_cnndm, linewidths=0.1, 
            annot=True, ax=ax[1], cmap=viridis, vmin=0, vmax=1, cbar=False);
ax[1].set_title('(b) CNN/DM', fontsize=15)
b.set_xticklabels(labels_cnndm, size = 15)
b.set_yticklabels(labels_cnndm, size = 15)

# XSum
b = sns.heatmap(pr_xsum, xticklabels=labels_xsum, yticklabels=labels_xsum, linewidths=0.1, 
            annot=True, ax=ax[2], cmap=viridis, vmin=0, vmax=1, cbar=True, cbar_ax=cbar_ax);
ax[2].set_title('(c) XSum', fontsize=15)
b.set_xticklabels(labels_xsum, size = 15)
b.set_yticklabels(labels_xsum, size = 15)

plt.tight_layout()
plt.savefig('./pretty_combined_pearson.jpg', dpi=300, bbox_inches='tight')
```
